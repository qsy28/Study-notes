# Matplotlib学习总结

## 通用设置

### 以折线图为例  plt.plot()

```python
from matplotlib import pyplot as plt

x = range(2,26,2)
y_1 = [15,13,14.5,17,20,25,26,26,27,22,18,15]
y_2 = [16,18,14.5,17,23,25,18,22,27,22,15,18]
```

```python
## 设置图片大小
plt.figure(figsize=(20,8),dpi=60)

## 显示中文
# 在 notebook 里，设置下面两行才能显示中文
plt.rcParams['font.family'] = ['sans-serif']
# 如果是在 PyCharm 里，只要下面一行即可
plt.rcParams['font.sans-serif'] = ['SimHei']
#设置坐标中的符号“-”
plt.rcParams['axes.unicode_minus'] = False

## 设置x,y轴刻度
plt.xticks(range(2,25))
#plt.yticks(range(min(y),max(y)+1))

# 设置字符串标度：数字与字符串一一对应，长度要一致
_x = list(x)
_xticks_labels = ["你好{}".format(i) for i in _x]
plt.xticks(_x,_xticks_labels,rotation = 90)#rotation表示顺时针旋转参数

## 添加描述
plt.xlabel("时间",size = 20)
plt.ylabel("温度",color = 'r')
plt.title("时间-温度折线图")

## 绘制网格
plt.grid(alpha = 0.5, linestyle = '-.') #alpha设置透明度


# 图片保存
#plt.savefig('./1.png')

# 绘制多主体信息，设置label、颜色、线条风格、线条粗细、透明度
plt.plot(x,y_1,label='自己',color = 'g',linestyle = '--',linewidth = 3, alpha = 0.5)
plt.plot(x,y_2,label='他人')

## 添加图例(要在plot()函数之后)
plt.legend(loc = "upper left")

plt.show()
```

## 绘制统计图

### 散点图 plt.scatter()

```python
from matplotlib import pyplot as plt

x = range(2,26,2)
y_1 = [15,13,14.5,17,20,25,26,26,27,22,18,15]
y_2 = [16,18,14.5,17,23,25,18,22,27,22,15,18]
```

```python
## 显示中文
# 在 notebook 里，设置下面两行才能显示中文
plt.rcParams['font.family'] = ['sans-serif']
# 如果是在 PyCharm 里，只要下面一行即可
plt.rcParams['font.sans-serif'] = ['SimHei']
#设置坐标中的符号“-”
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize = (20,8))

x_2 = range(32,56,2)
## 唯一区别

plt.scatter(x,y_1,label = '3月份')
plt.scatter(x_2,y_2,label = '10月份')

x_all = list(x)+list(x_2)
xticks_label_1 = ['3月{}日'.format(i) for i in x]
xticks_label_2 = ['10月{}日'.format(i-30) for i in x_2]
xticks_label = xticks_label_1 + xticks_label_2
plt.xticks(x_all,xticks_label,rotation = 45)

plt.legend()

plt.xlabel("时间")
plt.ylabel("销量")
plt.title("销量时间图")

plt.show()
```

### 条形图  plt.bar()&plt.barh()

```python
from matplotlib import pyplot as plt

## 显示中文
# 在 notebook 里，设置下面两行才能显示中文
plt.rcParams['font.family'] = ['sans-serif']
# 如果是在 PyCharm 里，只要下面一行即可
plt.rcParams['font.sans-serif'] = ['SimHei']
#设置坐标中的符号“-”
plt.rcParams['axes.unicode_minus'] = False
```

```python
movie = ['战狼','天气之子','猪头少年','我的英雄学院','鬼灭之刃','世纪缘','名侦探柯南']
money = [50,30,20,34,28,33,27]
```

```python
plt.bar(range(len(movie)),money,width=0.3)

#用xticks可以进行标签的旋转
plt.xticks(range(len(movie)),movie,rotation = 90)

plt.xlabel('电影')
plt.ylabel('票房（单位：亿元）')

plt.show()
```

```python
#绘制横向的条形图
plt.barh(range(len(movie)),money,height = 0.3,color = 'orange')
plt.yticks(range(len(movie)),movie)
plt.grid(alpha = 0.5)
plt.show()
```

```python
## 绘制复合条形图——需要加入偏移量

day_1 = [10,5,7,4,6.5,7.8,3.9]
day_2 = [11.5,4.5,5,5.7,4.5,6.7,3]
day_3 = [12.9,6.6,4.7,4.5,3.5,5,4.3]

bar_width = 0.2
x_1= list(range(len(movie)))
x_2 = [i+bar_width for i in x_1]
x_3 = [i+bar_width*2 for i in x_1]

plt.bar(x_1,day_1,width = bar_width,label = "第一天")
plt.bar(x_2,day_2,width = bar_width,label = "第二天")
plt.bar(x_3,day_3,width = bar_width,label = "第三天")

#设置x轴的刻度
plt.xticks(x_2,movie,rotation = 90)

plt.legend()

plt.show()
```
