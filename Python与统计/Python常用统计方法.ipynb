# 数值类描述统计

## Numpy

### 平均值、方差、标准差

```python
import numpy as np
```

```python
a = [5, 6, 16, 9]
np.mean(a)
```

```python
np.average(a, weights = [1, 2, 1, 1])
```

```python
np.ptp(a)
```

```python
np.var(a) # 计算总体方差
```

```python
np.var(a, ddof = 1) # 计算样本方差
```

```python
b = [[4, 5], [6, 7]]
b
```

```python
np.var(b) # 计算矩阵所有元素的方差
```

```python
np.var(b, axis = 0) # 计算矩阵每一列的方差
```

```python
np.var(b, axis = 1) # 计算矩阵每一行的方差
```

```python
np.std(a) # 计算总体标准差
# 矩阵运算同理np.var()
```

### 理解协方差矩阵

```python
b = [7,10,4,9]
np.var(b,ddof = 1)
```

```python
# cov计算一维向量或者二维矩阵

np.cov(a,b) #默认计算样本协方差矩阵

# 协方差矩阵：其中对角线元素是每个维度的方差，非对角线上的元素则是不同维度间的协方差。
```

```python
# np中矩阵默认行代表feature，我们想要了解的是feature的相关性，例如身高和体重

c = np.array([[1, 5, 6], [4, 3, 9], [4, 2, 9], [4, 7, 2]])
print(c)
```

```python
np.var([1,5,6],ddof=1)
```

```python
np.cov(c) #bias 默认为False,计算样本协方差
```

### 皮尔逊相关系数

```python
# 与np.cov()同理
np.corrcoef(c)
```
