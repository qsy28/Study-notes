## nan和inf

#### nan(NAN,Nan):not a number表示不是一个数字
什么时候numpy中会出现nan：
- 当我们读取本地的文件为float的时候，如果有缺失，就会出现nan
- 当做了一个不合适的计算的时候(比如无穷大(inf)减去无穷大)

#### inf(-inf,inf):infinity,inf表示正无穷，-inf表示负无穷
什么时候回出现inf包括（-inf，+inf）
- 比如一个数字除以0，（python中直接会报错，numpy中是一个inf或者-inf）

## 统计函数

- 求和：t.sum(axis=None)
- 均值：t.mean(a,axis=None)  受离群点的影响较大
- 最大值：t.max(axis=None) 
- 最小值：t.min(axis=None)
- 标准差：t.std(axis=None) 
- 极值：np.ptp(t,axis=None) 即最大值和最小值只差
- 中值：np.median(t,axis=None) 

```python
import numpy as np

t = np.arange(12).reshape((3,4)).astype("float")
t[1,2:] = np.nan
t
```

```python
# 按行求均值，填充nan
def fill_ndarray(t):
    for i in range(t.shape[1]):
        tem_col = t[:,i] #按列取
        nan_num = np.count_nonzero(tem_col != tem_col) #统计nan个数
        if nan_num != 0: #说明该列具有nan
            # 取列中不为nan的数组
            temp_not_nan_col = tem_col[tem_col == tem_col]
            # 将列中为nan的值赋值为不为nan数组的均值
            tem_col[np.isnan(tem_col)] = temp_not_nan_col.mean()
    return t


t = np.arange(12).reshape((3,4)).astype("float")
t[1,2:] = np.nan
t = fill_ndarray(t)
t
```

## copy和view

#### a和b相互影响
- a=b 完全不复制，a和b相互影响
- a = b[:],视图的操作，一种切片，会创建新的对象a，但是a的数据完全由b保管，他们两个的数据变化是一致的，
#### a和b相互不影响
- a = b.copy(),复制，a和b互不影响

```python
# a = b
b = np.random.randint(10,20,(4,5))
b
```

```python
a = b
a
```

```python
a[0,0] = 0
a
```

```python
b # b的值也跟着改变了
```

```python
# a = b[:]
b1 = np.random.randint(5,20,(3,5))
b1
```

```python
a1 = b1[1:,1:]
a1
```

```python
a1[0,0] = 100
a1
```

```python
b1 # b1的数值也跟着改变了
```

```python
# a = b.copy()
b2 = np.random.randint(5,20,(3,5))
b2
```

```python
a2 = b2.copy()
a2
```

```python
a2[0,0] = 100
a2
```

```python
b2 # b2的数值没有改变，a2，b2互不影响
```
