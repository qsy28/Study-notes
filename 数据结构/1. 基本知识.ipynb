$ a + b + c = 1000 , a^2 + b^2 = c^2 , a、b、c为自然数，把所有的a,b,c的所有组合拿出来。 $

```python
import time
```

```python
# 枚举法
start_time = time.time()
for a in range(0,1001):
    for b in range(0,1001):
        for c in range(0,1001):
            if a+b+c == 1000 and a**2 + b**2 == c**2:
                print("a,b,c:%d %d %d"%(a,b,c))
end_time = time.time()
print("time_cost:",end_time-start_time)
```

$ T = 1000*1000*1000*2 $

$ 如果 a+b+c = N, 那么 T = N^3 * 2 $

$时间复杂度：O(n^3)$

```python
start_time = time.time()
for a in range(0,1001):
    for b in range(0,1001):
        c = 1000 - a - b
        if a**2 + b**2 == c**2:
            print("a,b,c:%d %d %d"%(a,b,c))
end_time = time.time()
print("time_cost:",end_time-start_time)
```

$时间复杂度：O(n^2)$

## 2. 代码执行时间衡量

### python内置类型性能分析

```python
from timeit import Timer
```

```python
# append方法 
def test1():
    li = []
    for i in range(10000):
        li.append(i) #列表尾部添加
        
# insert
def test6():
    li = []
    for i in range(10000):
        li.insert(0,i) #insert() 函数用于将指定对象插入列表的指定位置。

# += 方法
def test2():
    li = []
    for i in range(10000):
        li += [i]
        
# + 方法
def test7():
    li = []
    for i in range(10000):
        li = li + [i]

# 列表生成器
def test3():
    li = [i for i in range(10000)]

    
# list转换
def test4():
    li = list(range(10000))
    
# extend
def test5():
    li = []
    for i in range(1000):
        li.extend([i])
```

```python
timer1 = Timer("test1()","from __main__ import test1")
print("append：",timer1.timeit(1000))

timer6 = Timer("test6()","from __main__ import test6")
print("insert：",timer6.timeit(1000))

timer2 = Timer("test2()","from __main__ import test2")
print("+=：" ,timer2.timeit(1000))

timer7 = Timer("test7()","from __main__ import test7")
print("+：",timer7.timeit(1000))

timer3 = Timer("test3()","from __main__ import test3")
print("列表生成器：",timer3.timeit(1000))

timer4 = Timer("test4()","from __main__ import test4")
print("list转换：",timer4.timeit(1000))

timer5 = Timer("test5()","from __main__ import test5")
print("extend：",timer5.timeit(1000))
```
