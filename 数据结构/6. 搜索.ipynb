### 搜索：在序列中找到某个元素是否存在

## 二分查找

#### 顺序表

#### 有序列表

#### 时间复杂度

- **最优时间复杂度：$O(1)$**


- **最坏时间复杂度：$O(\log n)$**




```python
def binary_search(alist,item):
    """二分查找,递归"""
    n = len(alist)
    if n>0:
        mid = n//2
        if item  == alist[mid]:
            return True
        elif item < alist[mid]:
            return binary_search(alist[:mid],item)
        else:
            return binary_search(alist[mid+1:],item)
    return False
```

```python
li = [13, 22, 22, 33, 45, 54, 69, 78,78]
binary_search(li,15)
```

```python
binary_search(li,78)
```

```python
def binary_search_2(alist,item):
    """二分查找，非递归"""
    n = len(alist)
    start = 0
    end = n-1
    while start <= end:
        mid = (start+end) // 2
        if item == alist[mid]:
            return True
        elif item < alist[mid]:
            end = mid -1
        else:
            start = mid + 1
    return False
```
