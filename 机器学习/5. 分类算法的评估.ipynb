### **3. 精确率（查的准——查准率）、召回率（查的全——查全率）、F1**

$$precision = \frac{TP}{TP + FP}$$

$$recall = \frac{TP}{TP + FN}$$

如果研究问题关注以上两个指标中的某一个，则可直接利用precision或者recall进行学习器选择，若需综合考虑两者，可通过：

- 以precision为纵轴，recall为横轴作图，就可以得到**P-R曲线** ：P-R曲线下的面积越大，表示学习器更加优秀


- 但由于面积估算不容易，提出了**平衡点 $BEP$ ** ，BEP是 $precision=recall$ 时的度量


- BEP过于简单，更常用的是**F1**，$F1$是$p$跟$r$的调和平均数

$$\frac{1}{F1} = \frac{1}{2}(\frac{1}{P} + \frac{1}{R})$$

- $F_{\beta}$是$p$跟$r$的加权调和平均数，${\beta}> 1$时，$R$有更大影响

$$\frac{1}{F_{\beta}} = \frac{1}{1+{\beta}^2}(\frac{1}{P} + \frac{{\beta}^2}{R})$$

一般会有多个二分类混淆矩阵（多次验证），有宏P，宏R，宏F1；微P，微R，微F1

宏是指计算每个混淆矩阵的P、R、F1，然后平均，微是指计算每个混淆矩阵的TP、FN、FP、TN，平均后计算P、R、F1

> 关注的是正例（可理解为更为关注的类别）的预测情况，对于多分类而言，计算的是各个分类的精确率与召回率

**`sklearn.metrics.classification_report(y_true, y_pred, target_names=None)`**

y_true：真实目标值

y_pred：估计器预测目标值

target_names：目标类别名称

return：每个类别精确率、召回率、F1、support(多少样本预测为该类)

### **4. ROC与AUC**

ROC曲线的纵轴时TPR（真正例率），横轴时FPR（假正例率）

$$ TPR = \frac{TP}{TP+FN} $$

$$ FPR = \frac{FP}{TN+FP} $$

ROC曲线下的面积为AUC,该值越大越好
